### One hot Encoding

```python
df_OneHot = pd.get_dummies(copy, columns=["workclass", "race", "native-country", "relationship", "marital-status", "education", "sex","occupation"])
df_OneHot.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import std
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
df_OneHot.reset_index(inplace=True)
df_OneHot
```

```python
x=df_OneHot.drop(['>50K, <=50K'],axis=1)
y=df_OneHot['>50K, <=50K']
```

```python
x.drop(['index'],axis=1,inplace=True)
x
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y,test_size=0.3,random_state=25, shuffle=True)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
```

# SVM

```python
svm_model = svm.SVC(probability=True)
svm_model.fit(xtrain, ytrain)
```

```python
pred_svm = svm_model.predict(xtest)
pred_svm[0:9]
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
pred_svm
```

```python
ytest
```

```python
ytest.head
```

```python
confusion_matrix(ytest, pred_svm)
```

```python
pred_prb_svm = svm_model.predict_proba(xtest)
pred_prb_svm[0:20, 0:20]
```

```python
tn, fp, fn, tp = confusion_matrix(ytest, pred_svm).ravel()
conf_matrix = pd.DataFrame({"pred_Survived": [tp, fp], "pred_Not Survived": [
                           fn, tn]}, index=["Survived", "Not Survived"])
conf_matrix
```

```python
accuracy = (tp + tn) / (tp + fp + tn + fn)
print("Accuracy: {}".format(accuracy))
```

# Losgitic regression

```python
lr_model = LogisticRegression(random_state=25)
lr_model.fit(xtrain, ytrain)
```

```python
pred = lr_model.predict(xtest)
pred[0:9]
```

```python
confusion_matrix(ytest, pred)
```

```python
pred_prb = lr_model.predict_proba(xtest)
pred_prb[0:20, 0:20]
```

```python
tn, fp, fn, tp = confusion_matrix(ytest, pred).ravel()
conf_matrix = pd.DataFrame({"pred_<=50K": [tp, fp], ">50K": [
                           fn, tn]}, index=["<=50k", ">50K"])
conf_matrix
```

```python
accuracy = (tp + tn) / (tp + fp + tn + fn)
print("Accuracy: {}".format(accuracy))
```

```python
test_list = confusion_matrix(ytest, pred).ravel()
index_list = confusion_matrix(ytest, pred).ravel().nonzero()
tn, fp, fn, tp =list(itemgetter(*index_list)(test_list))
conf_matrix = pd.DataFrame({"pred_<=50K": [tp, fp], ">50K": [
                           fn, tn]}, index=["<=50k", ">50K"])
conf_matrix
```

```python
accuracy = (tp + tn) / (tp + fp + tn + fn)
print("Accuracy: {}".format(accuracy))
```
